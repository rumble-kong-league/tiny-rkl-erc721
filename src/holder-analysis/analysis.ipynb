```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
snapshot = json.loads(open("snapshot.json").read())
```

```python
_ = plt.hist([x['numKongsOwned'] for x in snapshot], bins=range(1, 10))
```

```python
_ = plt.hist([x['numKongsOwned'] for x in snapshot], bins=20)
```

```python
count, bins_count = np.histogram([x['numKongsOwned'] for x in snapshot], bins=50)
pdf = count / sum(count)
cdf = np.cumsum(pdf)
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
_ = plt.legend()
```

```python
count, bins_count = np.histogram([x['numKongsOwned'] for x in snapshot if x['numKongsOwned'] < 10], bins=50)
pdf = count / sum(count)
cdf = np.cumsum(pdf)
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
_ = plt.legend()
```
